## Distinct powers
Problem 29

Consider all integer combinations of ab for 2 ≤ a ≤ 5 and 2 ≤ b ≤ 5:

    22=4, 23=8, 24=16, 25=32
    32=9, 33=27, 34=81, 35=243
    42=16, 43=64, 44=256, 45=1024
    52=25, 53=125, 54=625, 55=3125

If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125

How many distinct terms are in the sequence generated by ab for 2 ≤ a ≤ 100 and 2 ≤ b ≤ 100?

```python
# imports
import sys
sys.path.append('../../modules')

from maths import gcf
from sequences import Prime_Factorizations
import copy
from itertools import product
from collections import Counter
from primes import num_from_pf_counter
```

```python

PF = Prime_Factorizations()
```

```python
search_range = 5

def pf_reduced(n: int):
    pf = copy.copy(PF[n])
    alpha = gcf(*pf.values(),PF=PF)
    for k,v in pf.items():
        pf[k] = int(v/alpha)
    return num_from_pf_counter(pf), alpha


def checkables(a,b):
    reduced_pf, alpha = pf_reduced(a)
    return reduced_pf, alpha * b


store = {}
for n in range(2,search_range+1):
    i,alpha = pf_reduced(n)
    store[i] = alpha

store


# def are_equal(a,b,c,d):
#     return checkables(a,b) == checkables(c,d)



# for a,b,c,d in product(range(2,7),range(2,7),range(2,7),range(2,7)):
#     print(f"{a}^{b},{c}^{d}",are_equal(a,b,c,d), a**b == c**d)



```

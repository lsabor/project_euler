## Triangular, pentagonal, and hexagonal
Problem 45

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:  
Triangle 	  	Tn=n(n+1)/2 	  	1, 3, 6, 10, 15, ...  
Pentagonal 	  	Pn=n(3n−1)/2 	  	1, 5, 12, 22, 35, ...  
Hexagonal 	  	Hn=n(2n−1) 	  	1, 6, 15, 28, 45, ...  

It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.


Link: https://projecteuler.net/problem=45

Date solved:  
2022/07/10

```python
# imports
from maths import is_pentagonal_number, is_hexagonal_number
```

```python
def get_triangle_number_n(n):
    return int(n * (n+1) / 2)

def get_next_polygonal_number(i: int):
    while True:
        n = get_triangle_number_n(i)
        if is_pentagonal_number(n) and is_hexagonal_number(n):
            return n
        i += 1
 
i = 286 # the 285th triangle number is a polygonal number, we have to find next
get_next_polygonal_number(i)
```
